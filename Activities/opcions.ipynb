## Pregunta **3**

Fórmula per calcular la distància entre dos punts en l'espai 3D:

$$
d(p_1, p_2) = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + (z_2 - z_1)^2}
$$


### **A**:

```python
def calcular_distancia(pos1, pos2):
    distancia = math.sqrt((pos1 - pos2) ** 2 + (pos1 - pos2) ** 2)
    return distancia
```

### **B**:

```python
def calcular_distancia(pos1, pos2):
    distancia = math.sqrt((pos1.x - pos2.x) ** 2 + (pos1.y - pos2.y) ** 2 + (pos1.z - pos2.z) ** 2)
    return distancia
```
